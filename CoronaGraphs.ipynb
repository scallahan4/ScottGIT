```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

CV = {'State':['Massachusetts','Maine', 'New Hampshire', 'Vermont','Connecticut','Rhode Island'],
     'Count':[41,0,4,1,2,3],
     'Active':[40,0,4,1,2,3]}

mydf=pd.DataFrame(CV,columns = ['State','Count','Active'])

mydf.head()

%matplotlib inline


mydf.plot.bar()

xval=mydf.State
ypos=np.arange(len(xval))
plt.xticks(ypos, mydf.State, rotation=45, horizontalalignment='right')
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

CV = {'City':['Hubei, China','Iran', 'Italy', 'Spain'],
     'Count':[67786,11364,15113,4334],
     'Active':[13171,7891,13052,4019],
    'Deaths':[3062,514,1016,122]}


mydf=pd.DataFrame(CV,columns = ['City','Count','Active','Deaths'])

mydf.head()

%matplotlib inline


mydf.plot.bar()

xval=mydf.City
ypos=np.arange(len(xval))
plt.xticks(ypos, mydf.City, rotation=45, horizontalalignment='right')
```
