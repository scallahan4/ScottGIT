# Conditional statement 

```python
# if statement 
temp = int(input('enter your temperature: '))
dewpoint = int(input('enter your dewpoint:'))
if temp >= 75 and dewpoint >= 65: 
    print("Conditions too warm")
elif temp <= 50:
    print("Temp too cold!")
else: 
    print("Ideal Temperature")
```
